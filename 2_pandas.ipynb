````python
import pandas as pd

````

````python
# Create from dictionary
data = {
    "name" : ["Sumed", "Rahul", "Priya"],
    "age"  : [22, 45, 33],
    "bp"   : [119, 135, 108]
}

df = pd.DataFrame(data)
print(df)
````

df.head()        # Show first 5 rows

df.tail()        # Show last 5 rows

df.shape         # (rows, columns) count

df.info()        # Data types of each column

df.describe()    # Statistics summary

````python


data = {
    "name"  : ["r", "p", "k"],
    "marks" : [1, 2, 3],
    "grade" : ["A", "B", "C"]
}

df = pd.DataFrame(data)

# Print DataFrame
print(df)

# Print shape
print(df.shape)
````

pd.DataFrame()     # Open pd toolbox → pick DataFrame tool

```

---

### Real Life Analogy 🔧
```
pd          = Your toolbox
.           = Open the toolbox
DataFrame   = The specific tool you need

# ACTIONS — need brackets ()
print(df)           # does something → displays
df.head()           # does something → shows first 5 rows
df.describe()       # does something → calculates statistics
df.dropna()         # does something → removes empty rows

# VALUES — no brackets
df.shape            # just holds → (rows, columns)
df.columns          # just holds → column names
df.index            # just holds → row numbers
df.dtypes           # just holds → data types
```

---

### Simple Rule Forever 💡
```
Ask yourself — "Is this DOING something or just STORING something?"

DOING something  → needs ()
STORING something → no ()
