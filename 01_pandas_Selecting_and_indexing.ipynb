```
### Real Life Analogy 🔧
```

# ACTIONS — need brackets ()
print(df)           # does something → displays
df.head()           # does something → shows first 5 rows
df.describe()       # does something → calculates statistics
df.dropna()         # does something → removes empty rows

# VALUES — no brackets
df.shape            # just holds → (rows, columns)
df.columns          # just holds → column names
df.index            # just holds → row numbers
df.dtypes           # just holds → data types
```

---

### Simple Rule Forever 💡
```
Ask yourself — "Is this DOING something or just STORING something?"

DOING something  → needs ()
STORING something → no ()

## Creating DataFrame

````python

# Create a DataFrame from a dictionary

data = {
    'EmpID':  [101, 102, 103, 104, 105],
    'Name':   ['Sumed', 'Priya', 'Rahul', 'Neha', 'Arjun'],
    'Dept':   ['Data', 'HR', 'Data', 'Finance', 'Data'],
    'Salary': [60000, 45000, 75000, 55000, 80000],
    'Exp':    [3, 2, 5, 4, 6]
}
df = pd.DataFrame(data)
print(df)
````

## Exploring the DataFrame

````python
# Key methods to explore any DataFrame
df.head(2)       # first 2 rows
df.tail(2)       # last 2 rows
df.shape          # (rows, columns)
df.columns        # column names
df.dtypes         # data types of each column
df.info()          # summary of the whole df
df.describe()      # stats for numeric columns
````

## Selecting Data

````python
# Select a single column → returns a Series
df['Name']

# Select multiple columns → returns a DataFrame
df[['Name', 'Salary']]

# Select rows by index → .iloc (integer location)
df.iloc[0]        # first row
df.iloc[1:3]      # rows 1 and 2

# Select rows by label → .loc
df.loc[0, 'Name']  # row 0, Name column
````

````python
# Select multiple columns → returns a DataFrame
df[['Name', 'Salary']]
````

````python
# Select rows by index → .iloc (integer location)
df.iloc[0]        # first row
df.iloc[1:4]      # rows 1 and 3
````

````python
df.iloc[:, -1]      # ALL rows, last column
````

````python
# Select rows by label → .loc
df.loc[2, 'Dept']  # row 0, Name column
````

````python
# loc → "label location" → uses index labels & column names
# Syntax: df.loc[row_label, col_label]
# IMPORTANT: loc end is INCLUSIVE (unlike iloc)

df.loc[0]                      # row with index label 0
df.loc[0, 'Name']             # single value → 'Sumed'

````

````python
df.loc[0:2, 'Name':'Dept']    # rows 0,1,2 — cols Name,Dept
````

````python
df.loc[:, 'Salary']            # all rows, Salary column
````

````python
# Setting a custom index changes loc behavior
df2 = df.set_index('EmpID')
df2.loc[103]                   # now fetch by EmpID, not 0,1,2
df2.loc[103, 'Salary']         # → 75000
````

````python
# .at and .iat → for accessing a SINGLE cell only
# Faster than loc/iloc for single value access

df.at[2, 'Salary']    # label-based → 75000 (same as loc but faster)


# When would you use this?
# When looping over millions of rows — .at/.iat are faster
# For normal analysis, loc/iloc is fine
````

````python
df.iat[2, 3]           # position-based → 75000 (row 2, col 3)
````
