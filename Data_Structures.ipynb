```python
# Step 1 — Create one patient dictionary
patient1 = {
    "name" : "Chakma",
    "age"  : 28,
    "bp"   : 125,
}

# Step 2 — Create another patient
patient2 = {
    "name" : "Munna",
    "age"  : 29,
    "bp"   : 97,
}

# Step 3 — Store both in a list
patients = [patient1, patient2]

# Step 4 — Print each patient's name and bp
for patient in patients:
    print(patient["name"], "→ BP:", patient["bp"])
```
