## 📖 Topic 3 — Filtering & Conditions

1. What is filtering?


In real projects, you never work with the full dataset. You always need a subset — "show me only Data dept employees", "find customers who spent more than ₹10,000", "get all records from 2024". That's filtering.


Pandas does this using something called a Boolean Mask.

2. What is a Boolean Mask?


When you write a condition on a column, Pandas doesn't give you the data directly. It first creates a Series of True/False values — one for each row. Then it uses that to filter.

```python
import pandas as pd

data = {
    'Name':   ['Sumed', 'Priya', 'Rahul', 'Neha', 'Arjun'],
    'Dept':   ['Data', 'HR', 'Data', 'Finance', 'Data'],
    'Salary': [60000, 45000, 75000, 55000, 80000],
    'Exp':    [3, 2, 5, 4, 6]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Step 1: condition creates a boolean mask
mask = df['Salary'] > 60000
print(mask)

# Step 2: pass mask into df[] to filter rows
df[mask]

# One-liner (most common style in real code)
df[df['Salary'] > 60000]
```

```python
# One-liner (most common style in real code)
df[df['Salary'] > 60000]
```

## & > AND (both condition must be true)

```python
# & → AND (both conditions must be True)
df[(df['Dept'] == 'Data') & (df['Salary'] > 60000)]


# WHY parentheses around each condition?
# & and | have higher precedence than == > <
# Without () Python evaluates wrong — always wrap each condition
```

##  | > OR one condition must true

```python
# | → OR (at least one condition must be True)
df[(df['Dept'] == 'HR') | (df['Salary'] > 70000)]

```

## ~ NOT condition

```python
# ~ → NOT (inverts the condition)
df[~(df['Dept'] == 'Data')]   # everyone NOT in Data dept
```

## isin condition

```python
# isin() → checks if value belongs to a list
# Much cleaner than writing multiple OR conditions

# Instead of this messy code:
df[(df['Dept'] == 'Data') | (df['Dept'] == 'HR')]

```

```python
# Use this clean version:
df[df['Dept'].isin(['Data', 'HR'])]

```

```python
# NOT isin → use ~ to invert
df[~df['Dept'].isin(['Data', 'HR'])]  # everyone NOT in Data or HR
```

## Between  RANGE FILTERING

```python
# between() → checks if value falls in a range (inclusive)
# Cleaner than writing >= and <= separately

# Instead of:
df[(df['Salary'] >= 55000) & (df['Salary'] <= 75000)]

```

```python
# Use this:
df[df['Salary'].between(55000, 75000)]

```

```python
# Also works on experience
df[df['Exp'].between(3, 5)]
```

## Query SQL STYLE FILTERING

```python
# query() → write filter conditions as a string
# Looks and reads like SQL WHERE clause
# Great for readability, especially with multiple conditions

df.query("Salary > 60000")

```

```python
df.query("Dept == 'Data' and Salary > 60000")

```

```python
df.query("Exp >= 3 and Exp <= 5")

```

```python
# Use @ to reference a Python variable inside query
min_sal = 60000
df.query("Salary > @min_sal")
```

## Filtering + Selecting column together

```python
# Real world: filter rows AND select specific columns
# Most common pattern in actual data analysis work

# Get Name and Salary of Data dept employees with Exp > 3
df[
    (df['Dept'] == 'Data') & (df['Exp'] > 3)
][['Name', 'Salary']]

```

```python
# Same using loc — cleaner for combined row+col selection
mask = (df['Dept'] == 'Data') & (df['Exp'] > 3)
df.loc[mask, ['Name', 'Salary']]
```
